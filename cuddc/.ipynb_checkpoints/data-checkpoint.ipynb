```python
import numpy as np
import matplotlib.pylab as plt
import numpy.fft
```

```python
data=np.fromfile('../sdaa_cuddc/a.bin',dtype='float32').view('complex64')
nch=65536
lo_ch=1024
lo_f=lo_ch/2048*240
bw=60
data=data[:data.shape[0]//nch*nch]

waterfall=np.fft.fftshift(np.abs(np.fft.fft(data.reshape((-1,nch)), axis=1))**2)
plt.semilogy (np.mean(waterfall, axis=0))
```
